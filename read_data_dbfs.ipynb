```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("PySpark-Marathi-Learning") \
    .getOrCreate()
```

```python
data = [
    ("Pune", "Haveli", "Village1", 120),
    ("Pune", "Haveli", "Village2", 90),
    ("Pune", "Mulshi", "Village3", 60),
    ("Solapur", "Pandharpur", "Village4", 150),
    ("Solapur", "Pandharpur", "Village5", 110)
]

columns = ["District", "Taluka", "Village", "HouseConnections"]

df = spark.createDataFrame(data, columns)
```

```python
df.display()
```

```python
df_filtered = df.filter(df["HouseConnections"] > 100)
df_filtered.show()
```

```python
df.groupBy("District") \
  .sum("HouseConnections") \
  .show()
```

```python
from pyspark.sql.functions import col

df.groupBy("District") \
  .sum("HouseConnections") \
  .withColumnRenamed("sum(HouseConnections)", "TotalConnections") \
  .show()
```

```python
result_df = df.filter(df["HouseConnections"] > 100) \
    .groupBy("District", "Taluka") \
    .sum("HouseConnections") \
    .withColumnRenamed("sum(HouseConnections)", "TotalConnections")

result_df.show()
```
